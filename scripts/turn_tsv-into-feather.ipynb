```python
import pandas as pd
```

```python
# TSV파일을 읽어서 feather 파일로 저장

def turn_csv_to_feater(file_name_tsv, file_name_feather):    
    name_csv = f"./election_data/{file_name_tsv}.tsv"
    name_feather = f"./election_data_feather/{file_name_feather}"

    df = pd.read_csv(name_csv, sep='\t')
    df.to_feather(name_feather)
```

```python
turn_csv_to_feater("2024_지역구_longform", "2024_지역구.feather")
turn_csv_to_feater("2024_지역구_wideform_partyonly", "2024_지역구_정당.feather")
turn_csv_to_feater("2024_비례", "2024_비례.feather")
```
